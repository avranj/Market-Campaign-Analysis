IN this dataset we have founf out some unneccessary columns and from some columns we can create new columns:
e.g. from birth date column we can create AGE column-:

```python
# First we will separate date from timestamp as birthdate is already in datetime data-type

df["BirthDate"]
```

```python
df["BirthDate"].dt.date
```

```python
df['birthdate']=df["BirthDate"].dt.date
```

```python
df
```

```python
df['birthdate']=df['birthdate'].astype(str)
```

```python
df.info()
```

```python
from datetime import datetime,date
def age_calculator(born):
    today=date.today()
    born = datetime.strptime(born, "%Y-%m-%d").date() 
    return today.year - born.year - ((today.month,  today.day) < (born.month,  born.day))
    
    
    
df['Age']=df['birthdate'].apply(age_calculator)
df
```

```python
df['Age'].value_counts()
```

```python
print(df['Age'].mean())
print(df['Age'].mode())
print(df['Age'].min())
print(df['Age'].max())

```

In above operation we have seen that "Age is having pretty much amount of values which are below 15 and above 90"

```python
# Now we will analyse why age section is having high and low values
df[(df['Type']=="Business") & (df['Age']<=15)]
```

```python
df[df['Age']==5]
```

- From above operation we can say that age values which are less than 5 are of business (company) age, So we can use it in our analysis.
- Ages above 90 are individual ages so these may not be true we can either remove them or have them in our analysis.
